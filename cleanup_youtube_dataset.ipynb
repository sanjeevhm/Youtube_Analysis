**Cleanup of US dataset**

```python
youtube_df_us.head()
```

```python
type(youtube_df_us)
```

```python
youtube_df_us.shape
```

```python
youtube_df_us.drop(['channelId','channelTitle','thumbnail_link'],axis=1)
```

```python
youtube_df_us.head()
```

```python
youtube_df_us.shape
```

```python
youtube_df_us.isna().sum()
```

```python
youtube_df_us = youtube_df_us.dropna()
```

```python
youtube_df_us.isna().sum()
```

```python
youtube_df_us['time_of_day'] = pd.cut(pd.to_datetime(youtube_df_us.publishedAt).dt.hour,
       bins=[0, 6, 12, 18, 24],
       labels=['Night', 'Morning', 'Afternoon', 'Evening'],
       right=False,
       include_lowest=True)
```

```python
youtube_df_us['day_of_week'] = pd.DatetimeIndex(youtube_df_us['publishedAt']).day_name()
youtube_df_us.head()
```

```python
youtube_df_us['day_of_week'].value_counts()
```

```python
youtube_df_us['time_of_day'].value_counts()
```
