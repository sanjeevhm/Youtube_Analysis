```python
# formating publised date to date and time for US
df_us['publishedAt'] = pd.to_datetime(df_us['publishedAt'])
df_us['published_date'] = df_us['publishedAt'].dt.date
df_us['published_time'] = df_us['publishedAt'].dt.strftime('%H:%M:%S%z')


#formatting trending date and time for US
df_us['trending_date'] = pd.to_datetime(df_us['trending_date'])
df_us['trendingdate_date'] = df_us['trending_date'].dt.date
df_us['trendingdate_time'] = df_us['trending_date'].dt.strftime('%H:%M:%S%z')
```

```python
# statistical analysis of view count for US
df_us["view_count"].describe().apply(lambda x: format(x, 'f'))
```

```python
# statistical analysis of likes for US
df_us["likes"].describe().apply(lambda x: format(x, 'f'))
```

```python
# statistical analysis of dislikes for US
df_us["dislikes"].describe().apply(lambda x: format(x, 'f'))
```

```python
# statistical analysis of comment count for US
df_us["comment_count"].describe().apply(lambda x: format(x, 'f'))
```

```python
#adding category column for US
import json
with open("data/US_category_id.json") as f:
    categories = json.load(f)["items"]
category_dict = {}
for cat in categories:
    category_dict[int(cat["id"])] = cat["snippet"]["title"]
df_us['category_name'] = df_us['categoryId'].map(category_dict)

#top categories in US
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.countplot(y = "category_name", data=df_us, palette = "Set1").set(title="USA")
plt.show()
```

```python
# Top 10 categories in US
df_us['category_name'].value_counts().head(10)
```

```python
#adding category column for india
import json
with open("data/IN_category_id.json") as f:
    categories = json.load(f)["items"]
category_dict = {}
for cat in categories:
    category_dict[int(cat["id"])] = cat["snippet"]["title"]
df_in['category_name'] = df_in['categoryId'].map(category_dict)

#top categories in canada
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.countplot(y = "category_name", data=df_in, palette = "Set3")
plt.show()

# Top 10 categories in canada
df_in['category_name'].value_counts().head(10)
```

```python
df_list = [df_us, df_ca, df_in]
df_name_list = ['United States', 'Canada', 'India']
```

```python
def wordcld(a,j):
    from wordcloud import WordCloud
    import matplotlib.pyplot as plt
    title_words = list(a["title"].apply(lambda x: x.split() if x is not None else "None"))
    title_words = [x for y in title_words for x in y]
    #print(df_name_list[j])
    wc = WordCloud(width=1200, height=500, 
                                collocations=False, background_color="white", 
                                colormap="tab20b").generate(" ".join(title_words))
    plt.figure(figsize=(15,10))
    plt.imshow(wc, interpolation='bilinear')
    #plt.axis("off")
    plt.ylabel(df_name_list[j])
```

```python
j=0
while j<len(df_list):
    #print(df_name_list[j])
    wordcld(df_list[j],j)
    j=j+1
```

```python
#analysis of description USA
des = list(df_us["description"].apply(lambda x: x.split() if x else []))
des_count = [len(x) for x in des]
df_us['description_count'] = des_count

df_us['description_count'].value_counts().nlargest(10).plot(kind='barh')
plt.xlabel("No of Videos", labelpad=14)
plt.ylabel("Number of words in description", labelpad=14)
plt.title("Top 10 Videos vs Description word count: USA", y=1.02);
```
