```python
import numpy as np
import pandas as pd
```

```python
df_ca = pd.read_csv('/content/drive/MyDrive/Data mining/Project/categorised_youtube_dataframe_ca.csv',engine='python')
df_us = pd.read_csv('/content/drive/MyDrive/Data mining/Project/categorised_youtube_dataframe_us.csv',engine='python')
df_in = pd.read_csv('/content/drive/MyDrive/Data mining/Project/categorised_youtube_dataframe_in.csv',engine='python')
```

```python
trend_days_ca = df_ca.groupby(['video_id'])['video_id'].agg(total_trend_days=len).reset_index()
trend_days_ca.head()
```

```python
df_ca = pd.merge(df_ca, trend_days_ca, on='video_id')

df_ca.head()
```

```python
df_ca_last = df_ca.drop_duplicates(subset=['video_id'], keep='last', inplace=False)
df_ca_last.head()
```

```python
df_ca.count()
```

```python
trend_days_us = df_us.groupby(['video_id'])['video_id'].agg(total_trend_days=len).reset_index()
trend_days_us.head()

df_us = pd.merge(df_us, trend_days_us, on='video_id')

df_us.head()
df_us_last = df_us.drop_duplicates(subset=['video_id'], keep='last', inplace=False)
df_us_last.head()
```

```python
trend_days_in = df_in.groupby(['video_id'])['video_id'].agg(total_trend_days=len).reset_index()
df_in = pd.merge(df_in, trend_days_in, on='video_id')
df_in_last = df_in.drop_duplicates(subset=['video_id'], keep='last', inplace=False)
df_in_last.head()
```

```python
df_Views_Analysis = df_ca_last.groupby('category_name')["view_count"].mean().reset_index()
df_Like = df_ca_last.groupby('category_name')["likes"].mean().reset_index()
df_Dislike = df_ca_last.groupby('category_name')["dislikes"].mean().reset_index()
df_trending_days = df_ca_last.groupby('category_name')["total_trend_days"].mean().reset_index()
df_Comments = df_ca_last.groupby('category_name')["comment_count"].mean().reset_index()
df_Views_Analysis["Likes"] = df_Like.likes
df_Views_Analysis["Dislikes"] = df_Dislike.dislikes
df_Views_Analysis["TrendingDays"] = df_trending_days.total_trend_days
df_Views_Analysis
```

```python
df_Views_Analysis.plot(x = "category_name", y = "view_count", kind = "bar")
plt.show()
```

```python
df_Views_Analysis.plot(x = "category_name", y = "TrendingDays", kind = "bar")
plt.show()
```
